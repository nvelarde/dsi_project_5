### Feature Selection via SelectKBest

- Running SelectKBest using default k=10 -- select 10 'best' features

```python
kbest_dictionary = general_transformer(SelectKBest(k=10), scaled_dictionary)
```

```python
kbest_dictionary['processes']
```

```python
kbest_dictionary['transformer']
```

```python
# confirming K-Best transformer selected 10 features

pd.DataFrame(kbest_dictionary['X_train']).shape, pd.DataFrame(kbest_dictionary['X_test']).shape
```

```python
pd.DataFrame(kbest_dictionary['X_train']).head()
```

```python
pd.DataFrame(kbest_dictionary['X_test']).head()
```

```python
# mask for features -- True are the ones selected by KBest

kbest_feature_mask = kbest_dictionary['transformer'].get_support()
kbest_feature_mask
```

```python
kbest_feature_mask.shape
```

```python
features_df = pd.DataFrame(X.columns.tolist(), columns=["Feature"])
features_df.shape
```

```python
features_df.head()
```

```python
# Applying the KBest Feature Mask to get the features selected by KBest

kbest_selected_features_df = features_df[kbest_feature_mask == True]
kbest_selected_features_df.shape
```

```python
kbest_selected_features_df
```

## Building Model with Grid Search

### Logistic Regression

##### Tuning hyperparameter 'C'

```python
# Defining hyperparameter ranges

log_reg = LogisticRegression()

log_reg_params = {'C': np.linspace(0.005, 5, 500),
                  'penalty': ['l1'],
                 }
```

```python
log_reg_gs_dictionary = general_model(GridSearchCV(log_reg, 
                                                   log_reg_params, 
                                                   cv=10, 
                                                   verbose=1, 
                                                   n_jobs=-1), 
                                      kbest_dictionary)
```

```python
log_reg_gs_dictionary['processes']
```

```python
# Grid Search Results

log_reg_gs_results_df = pd.DataFrame(log_reg_gs_dictionary['model'].cv_results_)
log_reg_gs_results_df.head(5)
```

```python
log_reg_gs_results_df.shape
```

```python
log_reg_gs_dictionary['model'].best_estimator_
```

```python
log_reg_gs_dictionary['model'].best_params_
```

**Grid Search Best Score**

```python
# best test score

log_reg_gs_dictionary['model'].best_score_
```

```python
log_reg_gs_results_df['mean_train_score'].max()
```

```python
log_reg_gs_results_df[log_reg_gs_results_df['mean_test_score'] == log_reg_gs_results_df['mean_test_score'].max()] 
```

**Grid Search Salient Features**

```python
gs_features = pd.DataFrame({'coef': log_reg_gs_dictionary['model'].best_estimator_.coef_[0],
                            'feature': kbest_selected_features_df['Feature'].values
                           })

gs_features
```
